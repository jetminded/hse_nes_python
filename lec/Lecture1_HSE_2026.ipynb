#### Имя (идентификатор)

Имена переменных удовлетворяют следующим правилам:

- имя должно содержать только буквы, цифры и знаки нижнего подчеркивания (`_`)
- имя должно начинаться только с буквы или нижнего подчеркивания

**Примечания:**

1. Имена переменных чувствительны к регистру(case-sensetive) - `age`, `Age` и `AGE` являются тремя разными именами

3. Советую называть переменные осмысленно, чтобы другой программист смог понять за что именно отвечает та или иная переменная

#### Срезы

`строка[начало:конец:шаг]` - выбирает подстроку начиная с символа `начало` и до символа `конец`, **не включая** конец, с шагом `шаг`

````python
alphabet = "АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ"

slice_1 = alphabet[0:3]    # Символы с 0 по 2: 'АБВ'
slice_2 = alphabet[:5]     # С начала до 5-го символа: 'АБВГД'
slice_3 = alphabet[10:]    # С 10-го символа до конца: 'ЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ'
slice_4 = alphabet[::2]    # Каждый второй символ: 'АВДЁЗЙЛНРТФЦШЪЬЭЯ'
slice_5 = alphabet[::-1]   # Обратный порядок: 'ЯЮЭЬЫЪЩШЧЦХФУТСРПОНМЛКЙИЗЖЁЕДГВБА'

print("Первые три буквы:", slice_1)
print("Первые пять букв:", slice_2)
print("С десятой до конца:", slice_3)
print("Каждый второй символ:", slice_4)
print("Алфавит задом наперёд:", slice_5)
````

#### Индексация

Абсолютно аналогично строке. Можно считать, что строка - это список из символов

````python
my_list = ["яблоко", "банан", "вишня", "дыня", "инжир"]

# Получение элемента по индексу (положительному и отрицательному)
first = my_list[0]        # 'яблоко'
last = my_list[-1]        # 'инжир'
third_from_end = my_list[-3] # 'вишня'

print("Первый элемент:", first)
print("Последний элемент:", last)
print("Третий с конца:", third_from_end)

# Срезы (синтаксис: список[начало:конец:шаг])
slice1 = my_list[1:4]      # ['банан', 'вишня', 'дыня'] (с 1 по 3 индекс)
slice2 = my_list[:3]       # ['яблоко', 'банан', 'вишня'] (первые три)
slice3 = my_list[2:]       # ['вишня', 'дыня', 'инжир'] (с третьего до конца)
slice4 = my_list[::2]      # ['яблоко', 'вишня', 'инжир'] (каждый второй)
slice5 = my_list[::-1]     # ['инжир', 'дыня', 'вишня', 'банан', 'яблоко'] (реверс)

print("Срез [1:4]:", slice1)
print("Срез [:3]:", slice2)
print("Срез [2:]:", slice3)
print("Срез [::2]:", slice4)
print("Срез [::-1] (реверс):", slice5)
````

### NoneType

`NoneType` - специальный тип данных для объекта `None`:

#### Что такое `None`?

`None` - специальное значение переменной, означающее "ничего" или "отсутствие значения". Является аналогом `null` из других языков программирования.

Существует много случаев, когда следует использовать None.

Часто вы хотите выполнить действие, которое может работать либо завершиться неудачно. Используя `None`, вы можете проверить успех действия.

````python
x = None

x == True, x == False, x == None, x == 0, x == ""
````

````python
a = 500
b = 500

a is b, a == b # is - проверка на совпадение адресов. Под переменные a и b выделены разные места в памяти!
````

````python
a = None
b = None

a is b, a == b
# Значит, для None уже выделено место в памяти и когда создается переменная, то новое место не выделяется!
````

### Циклы: break, continue, else

#### continue

````python
# continue заканчивает исполнение текущей ИТЕРАЦИИ
# и сразу же переходит к следущей
````

````python
skip = 7

for n in range(10):
    if n == skip:
        print('*', end=', ')
        continue
    print(n, end=', ')
````

````python
data = [1, 2, 3, 4, 5, 6, 7]

while data:
    value = data.pop()  # удаляем последний элемент
    if 2 <= len(data) <= 4:
        continue
    print(value, end=', ')
````

#### break

````python
# break полностью прерывает исполнение цикла
````

````python
count = 0

for natural in [1, 2, 3, 4, -1, 20]:
    count +=1
    if natural < 0:
        print("ERROR")
        break


print(count)
````

````python
bad = [13, 5, 10]
errors = 0

n = 17
while n > 1:
    print(n, end=', ')
    n = n // 2 if n % 2 == 0 else n * 3 + 1

    if n in bad:
        print('ERROR', end='! ')
        errors += 1
        if errors == len(bad):
            print("FATAL ERROR - BREAK")
            break
````

````python
data = [1, 2, 3, 4, 5]

# Новичок
for i in range(len(data)):
    print(i, data[i], end=', ')

print()

# Опытный
for i, value in enumerate(data):
    print(i, value, end=', ')
````

````python
data = [1, 2, 3, 4, 5]

# Новичок
for value in data[::-1]:
    print(value, end=', ')

print()

# Опытный
for value in reversed(data):
    print(value, end=', ')
````

````python
list(reversed(data))
````

zip

````python
data = [1, 2, 3, 4, 5]
letters = ['a', 'b', 'c', 'd', 'e']

# Новичок
for i in range(len(data)):
    print(i, (data[i], letters[i]), end=', ')

print()

# Опытный
for value in zip(data, letters):
    print(value, end=', ')
````

````python
zip(data, letters)
````

# Немного заглянем под капот

````python
# All data in a Python program is represented by objects or by relations between objects

a = 1       # `имя` = `значение`. Имени `a` присваивается ссылка на объект со значением 1
A = id(a)   # `id`. Ссылка, где лежит объект в памяти
t = type(a) # `тип`. Какими свойствами обладает
a, A, t
````

````python
b = 1
c = 2
a = b + c   # Операция над объектами
````

````python
id(b), id(c), id(a)
````

````python
# Иммутабельные - не меняют значение по ссылке. IM = отрицание = НЕ
# Примеры: int, float, str, tuple, frozenset
a = 1
print(id(a))
b = a # b пока что ссылается на ту же ячейку, что и а
print(id(b))
a = a + 2
id(a), id(b)
````

````python
# Мутабельные - меняют значение по ссылке
# Примеры: list, set, dict
a = [1]
print(id(a))
b = a.copy()
print(id(b))
a += [2]
a, b
````
