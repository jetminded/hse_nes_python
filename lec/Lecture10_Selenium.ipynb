## Webdriver

- Выполнять действия (клики, ввод текста, нажатие клавиш).

## Поиск элементов (Locators)

Selenium предоставляет различные способы поиска элементов на странице с помощью методов `find_element` и `find_elements`. Основные локаторы:
https://htmlacademy.ru/blog/css/selectors


- **By.ID** — поиск по уникальному идентификатору элемента.
- **By.NAME** — поиск по атрибуту `name`.
- **By.CLASS_NAME** — поиск по классу элемента.
- **By.TAG_NAME** — поиск по тегу (например, `div`, `a`).
- **By.CSS_SELECTOR** — поиск по CSS-селектору.
- **By.XPATH** — поиск по XPath.

````python
from selenium import webdriver
from selenium.webdriver.common.by import By

# Инициализация драйвера
driver = webdriver.Chrome()

# Открытие страницы
driver.get("https://www.google.com")


# Поиск элемента по ID

element_by_id = driver.find_element(By.ID, "tophf")

# Поиск элемента по имени
element_by_name = driver.find_element(By.NAME, "referrer")

# Поиск элемента по классу
element_by_class = driver.find_element(By.CLASS_NAME, "pHiOh")

# Поиск элемента по тегу
element_by_tag = driver.find_element(By.TAG_NAME, "div")
````

````python
# Поиск элемента по CSS-селектору
element_by_css = driver.find_element(By.CSS_SELECTOR, "div") # Ищем по тегу

element_by_css = driver.find_element(By.CSS_SELECTOR, "div.LX3sZb") # Ищем по тегу + классу

element_by_css = driver.find_element(By.CSS_SELECTOR, ".LX3sZb") # Ищем по классу

element_by_css = driver.find_element(By.CSS_SELECTOR, "#tophf") # Ищем по id
````

## WebElement

**WebElement** представляет собой элемент на веб-странице, с которым **Selenium** может взаимодействовать. Это может быть кнопка, ссылка, текстовое поле и любой другой элемент на странице. Каждый WebElement обладает методами для взаимодействия, такими как клик `click()`, ввод текста `send_keys()`, получение текста элемента text и многими другими.

Пример взаимодействия с WebElement:

````python
driver = webdriver.Chrome()
driver.get("https://www.google.com")

element = driver.find_element(By.NAME, 'q')    # Находим элемент по его имени
element.send_keys('Selenium')  # Вводим текст в текстовое поле
element.submit()  # Отправляем форму
````

## Ожидания (Waits)

Веб-страницы могут загружаться с задержкой, поэтому Selenium предоставляет механизмы ожидания:

 - Неявное ожидание (Implicit Wait) — задает время, в течение которого WebDriver будет ждать появления элемента перед тем, как выбросить исключение.

- Явное ожидание (Explicit Wait) — позволяет ждать выполнения определенного условия (например, появление элемента, кликабельность элемента).

Пример:

````python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome()
driver.get("https://www.google.com")

# Неявное ожидание (ждем до 10 секунд)
driver.implicitly_wait(10)

# Ожидание видимости строки поиска Google
try:
    logo = WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.NAME, "q")))
    print("Строка Google видна!")
except:
    print("Строка Google не найдена или не видна")
finally:
    driver.quit()
````

## Работа с элементами

После того как элемент был найден с помощью локатора, Selenium позволяет выполнять с ним различные действия, такие как:

- Ввод текста `element.send_keys("текст")`
- Клики `button.click()`
- Очистка поля `element.clear()`
- Получение текста элемента `element.text`
- Получение атрибутов элемента

        ```
        link = driver.find_element(By.TAG_NAME, "a")
        print("Ссылка:", link.get_attribute("href"))
        ```

- Проверка состояния элемента
        -  Проверка видимости `element.is_displayed()`
        - Проверка активности `element.is_enabled()`
        - Проверка выбора элемента `element.is_selected()`
- Перетаскивание (Drag and Drop)

        ```
        from selenium.webdriver import ActionChains

        actions = ActionChains(driver)
        actions.drag_and_drop(source_element, target_element).perform()
        ```
- Наведение курсора на элемент - Используется для наведения курсора на элемент, чтобы вызвать всплывающие подсказки или меню `actions.move_to_element(element).perform()`

## Управление браузером

Selenium позволяет управлять браузером:

- Открыть новую вкладку — `driver.execute_script("window.open('');")`

- Переключиться между вкладками — `driver.switch_to.window(driver.window_handles[1])`

- Обновить страницу — `driver.refresh()`

- Перейти назад/вперед — `driver.back()`, `driver.forward()`

- Закрыть вкладку — `driver.close()`

- Закрыть браузер — `driver.quit()`

````python
driver = webdriver.Chrome()
driver.get("https://www.google.com")
driver.back()

# Открытие новой вкладки
driver.execute_script("window.open('https://www.example.com');")
````

````python
# Переключение на новую вкладку
driver.switch_to.window(driver.window_handles[1])
````

````python
# Закрытие вкладки
driver.close()
````

````python
# Возврат к первой вкладке
driver.switch_to.window(driver.window_handles[0])
````

````python
driver.quit()
````

## Управление cookies

Selenium позволяет работать с cookies:

- Добавить cookie — `driver.add_cookie({"name": "test", "value": "123"})`.`

- Получить cookie — `driver.get_cookie("name")`.`

- Удалить cookie — `driver.delete_cookie("name")`.`

````python
driver = webdriver.Chrome()
driver.get("https://www.google.com")

# Добавление cookie
driver.add_cookie({"name": "session", "value": "12345"})

# Получение cookie
print(driver.get_cookie("session"))

# Удаление cookie
driver.delete_cookie("session")
````

## Прокрутка

````python
from selenium.webdriver.common.action_chains import ActionChains

driver = webdriver.Chrome()
driver.get("https://doodles.google/")

# Использование ActionChains для прокрутки на заданное количество
actions = ActionChains(driver)
actions.scroll_by_amount(delta_x=50, delta_y=1000).perform()
````

````python
news = [item for item in driver.find_elements(By.CSS_SELECTOR, ".main-news-card__title")]
news
````

````python
titles = [new.text for new in news]
````

````python
news_title_links = [item.find_element(By.TAG_NAME, "a").get_attribute("href") for item in news]
````

````python
news_title_links
````

````python
import time
my_news = dict()
for title, url in zip(titles, news_title_links):
    try:
        driver.get(url)
        text_element = driver.find_element(By.CLASS_NAME, "post__text")
        my_news[title] = text_element.text
        print(title)
    except Exception as e:
        print("Error") # на случай если ссылка ведет куда-то во вне
        time.sleep(5)
````
