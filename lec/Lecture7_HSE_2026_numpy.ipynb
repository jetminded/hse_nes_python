## Библиотека NumPy
NumPy — библиотека Python, которую применяют в вычислениях, связанных с большими массивами данных. Её главное преимущество — скорость. Библиотека написана на языке C, который работает намного быстрее, чем обычный Python-код.

Многие задачи, с которыми ты столкнёшься, будут включать работу с табличными данными или матрицами. Модуль NumPy как раз и создан для работы с многомерными массивами, что делает его незаменимым инструментом для любого аналитика или инженера, работающего с машинным обучением.

[Документация NumPy](https://numpy.org/doc/)

**Преимущества NumPy:**

- **Быстрее** - NumPy использует алгоритмы, написанные на C, которые работают *значительно быстрее* чем те, что написаны на Python.
- **Меньше циклов** - NumPy помогает уменьшить количество циклов в коде вашей программы, что мешает не запутаться в них.
- **Чище код** - без вложенных циклов ваш код больше будет намного чище и проще для чтения.
- **Better quality** - код библиотеки написан сообществом профессионалов, поэтому он оптимизирован, приятен в работе и не содержит багов.


По этим причинам NumPy является де-факто **стандартом для работы с многомерными массивами** в анализе данных. Многие популярные библиотеки в своем исходном коде используют NumPy.

## Двумерный массив (матрица)

Когда ты слышишь слово «матрица», представляй себе таблицу данных со строками и столбцами.

> Матрицы в NumPy — это двумерные массивы, состоящие из строк и столбцов. С ними удобно выполнять математические операции, обрабатывать табличные данные и работать с изображениями.

Пример матрицы $A$ с размерностью $3 \times 2$:

$$
A = \begin{pmatrix}
1 & 2 \\
3 & 4 \\
5 & 6
\end{pmatrix}.
$$

Размерность $3 \times 2$ означает, что у этой матрицы три строки и два столбца.

````python
A = np.array([[1, 3, 5], [2, 4, 6]])
print(A)
````

Ключевые параметры массивов NumPy: размерность, форма и длина массива.

````python
print('Размерность:', A.ndim)  # Размерность (количество осей)
print('Форма:', A.shape)       # Форма (количество строк и столбцов)
print('Длина:', len(A))        # Длина (количество строк)
````

Количество измерений соответствует количеству осей.

  **Ось** — это направление, вдоль которого происходит изменение значений. Для двумерного массива:
- Ось 0 (первая ось) соответствует строкам.
- Ось 1 (вторая ось) соответствует столбцам.

## Двумерный массив



```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
```
Вот как выглядит этот массив:

```python
[[1 2 3]
 [4 5 6]]
```
У двумерного массива две оси:

- Ось 0 (оси строк): это `[1, 2, 3]` и `[4, 5, 6]` — строки массива.

- Ось 1 (оси столбцов): это `[1, 4]`, `[2, 5]`, `[3, 6]` — столбцы массива.

- Размерность массива: `(2, 3)` (2 строки и 3 столбца).

## Параметр axis


Параметр `axis` указывает, по какой оси ты хочешь выполнить операцию. Например, если нужно сложить элементы по строкам или столбцам, ты указываешь, по какой оси это надо сделать.

**Пример.** Данные о продажах продуктов в разных регионах нужно организовать в виде двумерного массива. Cтроки будут представлять регионы, а столбцы — разные продукты.

Здесь:

- Строки — это регионы: Регион 1, Регион 2, Регион 3.
- Столбцы — это продукты: Продукт A, Продукт B, Продукт C.

````python
sales_data = np.array([
    [100, 200, 150],  # Регион 1
    [120, 250, 130],  # Регион 2
    [90, 180, 160]    # Регион 3
])
````

Допустим, нужно узнать общую сумму продаж по всем продуктам и регионам. Для этого можно использовать метод `sum()` без указания оси.

````python
total_sales = sales_data.sum()
print(f'Общая сумма продаж: {total_sales}')
````

Чтобы узнать, сколько всего было продано каждого продукта, нужно сложить данные по столбцам — по оси 0. Для этого укажем параметр `axis=0`:

````python
product_sales = np.sum(sales_data, axis=0)
print(f'Сумма продаж по продуктам: {product_sales}')
````

### Генерация массивов

#### np.arange

Функция `numpy.arange(start, stop, step)` позволяет сгенерировать последовательность чисел в заданном интервале. В зависимости от того, сколько аргументов ты передаёшь, можно получить разные результаты:

- `arange(stop)`: создаёт последовательность чисел от `0` до `stop` (не включая `stop`).
- `arange(start, stop)`: создаёт последовательность от `start` до `stop` (не включая `stop`).
- `arange(start, stop, step)`: создаёт последовательность от `start` до `stop` с шагом `step` (не включая `stop`).

Это работает почти так же, как встроенная функция Python `range()`, но разница в том, что `arange()` возвращает массив типа `ndarray`, а не объект `range`.

````python
np.arange(0, 5, 0.5)
````

#### np.linspace

Функция `numpy.linspace(start, stop, num=50, endpoint=True)` позволяет сгенерировать указанное количество (`num`) равномерно распределённых значений в диапазоне от `start` до `stop`. При этом:

  - Параметр `num` задаёт, сколько чисел ты хочешь получить (по умолчанию 50).
  - Параметр `endpoint` определяет, включать или исключать конечную точку. Если `True`, то `stop` будет включён в последовательность; если `False`, то не будет.

#### np.logspace

Функция `numpy.linspace(start, stop, num=50, endpoint=True)` -аналогично linspace, но по логарифмической шкале

Можно указать `base` - основание логарифма

````python
np.logspace(0, 9, 10, base=2)
````

## Базовые типы данных

В NumPy массивы могут содержать элементы разных типов данных, и это один из ключевых моментов при работе с числами и строками. Чтобы управлять этими типами, используется параметр `dtype`, который определяет, какого типа будут элементы в массиве.

>**Важно.** Тип данных влияет на то, как числа будут храниться и отображаться. Например, если это целые числа (`int32` или `int64`), то они будут показаны без дробной части — просто 2. А если это числа с плавающей запятой (`float32` или `float64`), то они будут отображаться с точкой, как `2.0`.

##### Таблица распространённых типов данных в NumPy:
| Тип данных    | Описание                              | Пример отображения |
|---------------|---------------------------------------|--------------------|
| `int32`       | Целое число 32 бита                   | 2                  |
| `int64`       | Целое число 64 бита                   | 2                  |
| `float32`     | Число с плавающей точкой 32 бита      | 2.0                |
| `float64`     | Число с плавающей точкой 64 бита      | 2.0                |
| `bool`        | Логическое значение (True/False)      | True               |
| `str`         | Строка                               | 'example'          |


## Приведение типов


### Метод `astype()`

В NumPy есть метод `astype()`, который создаёт новый массив с другим типом данных. При этом исходный массив не изменяется — создаётся копия массива с новым типом данных.

Синтаксис:

```python
new_array = original_array.astype(dtype)
```

- **`dtype`** — это желаемый тип данных для нового массива, который может быть задан как строка (например, `'int32'`, `'float64'`, `'str'`) или как объект типа NumPy (например, `np.int32`, `np.float64`).

````python
# Исходный массив с плавающей точкой
arr_float = np.array([1.5, 2.3, 3.9])
print('Исходный массив:', arr_float)

# Приведение к целому типу
arr_int = arr_float.astype(np.int32) # Cоздаёт новый массив. Оригинальный массив остаётся неизменным
print('После приведения к целому типу:', arr_int)
````

**Замечание:** Если попытаться преобразовать несовместимые данные (например, строку, содержащую буквы, в целое число), то NumPy выдаст ошибку.

**Замечание:** При присваивании значения элементу массива приведение типов не происходит.

````python
arr_int[0] = 4.5
arr_int
````

## Методы изменения формы массива

### Метод `reshape`
Функция `numpy.reshape` изменяет форму массива. При этом меняется только его структура, а данные внутри массива остаются без изменений.

**Синтаксис:**
```python
numpy.reshape(a, shape)
```
- `a` — это массив для преобразования.
- `shape` — это новая форма массива (одно число или кортеж чисел). Одно из измерений может быть указано как `-1`, тогда NumPy автоматически вычислит его на основе остальных размеров.

>**Важно.** Форма массива — это набор его размеров по осям. Например, если преобразовываешь одномерный массив в матрицу, нужно указать количество строк и столбцов в новом массиве.

**Пример.** Преобразуем одномерный массив в матрицу. Есть данные о продажах за 36 месяцев (3 года), хранящиеся в виде одномерного массива. Но для удобства анализа нужно организовать их по годам и месяцам. Для этого можно использовать `reshape`.



````python
# Данные продаж за 36 месяцев (3 года)
sales_data = np.array([100, 120, 90, 130, 150, 140, 160, 170, 180, 200, 190, 210,
                       110, 130, 120, 140, 160, 150, 170, 180, 190, 210, 200, 220,
                       120, 140, 130, 150, 170, 160, 180, 190, 200, 220, 210, 230])

# Преобразуем одномерный массив в матрицу 3x12 (3 года по 12 месяцев)
reshaped_sales = sales_data.reshape(3, 12)

print(reshaped_sales)
````

### Метод `hstack`
Если нужно объединить несколько массивов, то поможет функция `numpy.hstack`. Она позволяет объединить массивы по горизонтали, то есть по столбцам. Это эквивалентно объединению вдоль второй оси, за исключением одномерных массивов, где объединение происходит вдоль первой оси.

**Синтаксис**:
```python
numpy.hstack(tup, dtype=None, casting='same_kind')
```
- `tup` — это кортеж массивов, которые ты хочешь объединить.
- `dtype` — тип данных для нового массива (необязательный параметр).
- `casting` — параметр, который определяет, как будут преобразовываться типы данных при объединении массивов.
 Возможные значения:
   - `'no'` — запрещает преобразование типов;
   - `'equiv'` — требует, чтобы типы были эквивалентны;
   - `'safe'` — допускает только безопасные преобразования;
   - `'same_kind'` — совершает преобразования между совместимыми типами;
   - `'unsafe'` — разрешает любые преобразования, даже если они могут привести к потере данных.
   
**Пример.** Есть два набора данных: один содержит информацию о продажах по месяцам за два года, а второй — данные о возвратах товаров за те же периоды. Чтобы провести анализ, удобно объединить эти данные в одну таблицу.

````python
sales = np.array([[100, 120, 90, 130, 150, 140, 160, 170, 180, 200, 190, 210],
                  [110, 130, 120, 140, 160, 150, 170, 180, 190, 210, 200, 220]])

# Возвраты по месяцам за те же 2 года
returns = np.array([[5, 3, 2, 4, 6, 5, 7, 6, 5, 4, 3, 5],
                    [4, 5, 3, 4, 6, 5, 8, 7, 6, 5, 4, 6]])

# Объединяем данные горизонтально
combined_data = np.hstack((sales, returns))

print(combined_data)
````

#### np.randint

**Синтаксис**:
```python
np.random.randint(low, high=None, size=None, dtype=int, endpoint=False)
```
- **`low`** (обязательно) — нижняя граница диапазона (включительно). Это минимальное значение, которое может быть сгенерировано.
- **`high`** (необязательно) — верхняя граница диапазона (исключительно). Если не указано, то генерируются числа от 0 до `low`.
- **`size`** (необязательно) — размер возвращаемого массива. Может быть указано как целое число (для одномерного массива) или кортеж (для многомерного массива).
- **`dtype`** (необязательно) — тип данных результата, по умолчанию `int`. Можно выбрать другой тип для результата.
- **`endpoint`** (необязательно) — если `True`, то верхняя граница (`high`) будет включительной.

````python
import numpy as np

# Генерация случайного количества проданных товаров за 10 дней (от 1 до 10)
sales = np.random.randint(1, 11, size=10)

# Генерация массива цен товаров (от 100 до 500)
prices = np.random.uniform(100, 500, size=10)

sales, prices
````

### Логические операции

Логические операции в NumPy позволяют сравнивать массивы и проверять условия для каждого элемента массива. Эти операции возвращают булевы массивы, где каждый элемент может быть либо `True` (истина), либо `False` (ложь). Булевы массивы могут быть использованы для фильтрации данных, выполнения различных анализов и других задач.

````python
predicted_sales = np.array([100, 150, 200])
actual_sales = np.array([90, 160, 190])

# Сравнение поэлементно с помощью оператора «меньше»
sales_below_forecast = actual_sales < predicted_sales
print(sales_below_forecast)  # array([ True, False,  True])
````

**Замечание:** как узнать количество True?

````python
print(sum(sales_below_forecast)) # приведение типов
````

#### Маски

**Пример.** Есть массив данных, в котором в результате округления или ошибок появились ложные отрицательные значения. Их нужно заменить на нули.


````python
a = np.array([-5, 2, 1, -4])
print('Исходный массив:', a)

# Создание маски для отрицательных элементов
mask = a < 0
print('Маска для отрицательных элементов:', mask)

# Замена отрицательных элементов на 0
a[a < 0] = 0
print('Результат:', a)
````

**np.logical_or**

Функция np.logical_or(a, b) выполняет логическую операцию «или» поэлементно для двух массивов. Если хотя бы один из соответствующих элементов равен True (или не ноль), результат будет True. Если оба элемента ложные (нулевые), результат будет False.



````python
# Данные о покупках: 1 — купил, 0 — не купил
customers_electronics = np.array([1, 0, 1, 1, 0])  # Покупка электроники
customers_sports = np.array([0, 1, 1, 0, 0])       # Покупка спортивных товаров

# Применение логической операции «или»
interested_customers = np.logical_or(customers_electronics, customers_sports)

print(interested_customers)
````

````python
customers_electronics | customers_sports # поэлементное ИЛИ
````

````python
a | b
````

## Агрегации

Агрегационные функции в NumPy позволяют легко и быстро получать ключевые статистические показатели по данным. Это полезно для анализа больших наборов данных, таких как продажи, измерения или другие числовые данные.

Для примера рассмотрим компанию, которая продаёт продукты питания. Нужно рассчитать основные статистические метрики для данных о продажах: **минимум**, **максимум**, **среднее**, **медиану** и **квантили**.

### Минимум и максимум


Для нахождения минимального и максимального значений в массиве используются функции `numpy.min()` и `numpy.max()`.

**Синтаксис:**
```pytnon
numpy.min(a, axis=None, out=None)
numpy.max(a, axis=None, out=None)
```
- `a` — входной массив.
- `axis` (опционально) — ось, вдоль которой выполняется операция.
- `out` (опционально) — массив, в который записывается результат.

````python
# Пример данных о продажах
sales_data = np.array([150, 200, 220, 180, 300, 150, 250, 400, 300, 350])

# Минимум и максимум
minimum = np.min(sales_data)
maximum = np.max(sales_data)

print('Минимум:', minimum)
print('Максимум:', maximum)
````

### Среднее значение


Среднее арифметическое — это сумма всех элементов, делённая на их количество. Оно показывает типичное значение в наборе данных.

**Синтаксис:**
```python
numpy.mean(a, axis=None, dtype=None, out=None)
```
  - `a` — входной массив.
  - `axis` (опционально) — ось, вдоль которой выполняется операция.
  - `dtype` (опционально) — тип данных для результата.
  - `out` (опционально) — массив для записи результата.

````python
mean_value = np.mean(sales_data)
print("Среднее:", mean_value)
````

### Медиана


Медиана — это значение, которое разделяет упорядоченный массив на две равные части: половина значений больше медианы, а другая половина меньше.

**Синтаксис:**
```python
numpy.median(a, axis=None, out=None)
```
  - `a` — входной массив.
  - `axis` (опционально) — ось, вдоль которой выполняется операция.
  - `out` (опционально) — массив для записи результата.  

````python
median_value = np.median(sales_data)
print('Медиана:', median_value)
````

### Квантили


**Синтаксис:**
```python
numpy.percentile(a, q, axis=None, out=None)
```
  - `a` — входной массив.
  - `q`— значение процентиля (от 0 до 100).
  - `axis` (опционально) — ось, вдоль которой выполняется операция.
  - `out` (опционально) — массив для записи результата.  

````python
# Первый и третий квартиль (25% и 75%)
first_quartile = np.percentile(sales_data, 25)
third_quartile = np.percentile(sales_data, 75)

print('25-й процентиль:', first_quartile)
print('75-й процентиль:', third_quartile)
````

### Сортировка элементов


Функция `numpy.sort()` возвращает отсортированную копию массива. Можно сортировать элементы массива как по возрастанию, так и по убыванию.

**Синтаксис:**
```python
numpy.sort(a, axis=-1, kind='quicksort', order=None, stable=None)
```

- `a` — массив, который нужно отсортировать.
- `axis` (опционально) — ось для сортировки. Если `None`, массив преобразуется в одномерный перед сортировкой (то есть все его элементы складываются в один ряд). По умолчанию — `-1`: массив сортируется по последней оси.
- `kind` (опционально) — алгоритм сортировки.
- `order` (опционально) — поля для сравнения. Задаётся, если массив содержит определённые поля и является не массивом чисел, а, например, массивом `tuple`, который можно сравнивать и сортировать только по первому элементу.
- `stable` (опционально) — если `True`, сохраняет относительный порядок равных значений. По умолчанию — `None`.

**Примечание.** Значения NaN сортируются в конец массива.

````python
a = np.array([5, 2, 244, 1])

# Сортировка по возрастанию
sorted_a = np.sort(a)
print('Отсортированный массив:', sorted_a)

# Сортировка по убыванию (разворот массива)
reversed_arr = sorted_a[::-1]
print('Массив по убыванию:', reversed_arr)
````

### Поэлементное сравнение: `numpy.maximum()` и `numpy.minimum()`

Функции `numpy.maximum()` и `numpy.minimum()` возвращают поэлементно максимальные и минимальные значения между двумя массивами или между массивом и скаляром.

<!-- Функция `np.maximum()` возвращает поэлементное максимальное значение между двумя массивами или между массивом и скаляром. Если входные данные имеют разную форму, NumPy автоматически применяет правила скейлинга к ним (broadcasting). Аналогично работает `np.minimum()`. -->
**Синтаксис:**
```python
numpy.maximum(x1, x2, out=None)
numpy.minimum(x1, x2, out=None)
```
- **`x1`, `x2`** — входные массивы, которые необходимо сравнить.
- **out** (опционально) — массив, в который будет записан результат. Если не указан, будет создан новый массив.

````python
def adjust_prices(store_prices, min_price, max_price):
    # Ограничиваем цены с помощью np.maximum и np.minimum
    adjusted_prices = np.minimum(np.maximum(store_prices, min_price), max_price)
    return adjusted_prices

# Пример использования
store_prices = np.array([8.99, 15.50, 22.99, 5.49, 30.00])
min_price = 10.00
max_price = 25.00

adjusted_prices = adjust_prices(store_prices, min_price, max_price)
print(adjusted_prices)
````

### Скалярное произведение векторов

Рассмотрим два вектора $a$ и $b$ в n-мерном пространстве  
$a = (a_1, a_2, a_3, \dots a_n)$   
$b = (b_1, b_2, b_3, \dots b_n)$   
Скалярное произведение векторов $a$ и $b$ определяется следующим образом:  
$$\langle a, b \rangle = a_1 b_1 + a_2 b_2 + a_3 b_3 \dots + a_n b_n = \sum_{i = 1}^{n} a_i b_i$$

````python
a = np.array([3, 1, 5, 2])
b = np.array([2, 5, 2, 4])
# <a, b> = 3*2 + 1*5 + 5*2 + 2*4

# 3 варианта написания:
print(a @ b)    # python 3 style
print(a.dot(b))
print(np.dot(a, b))
````

### Умножение матриц
  
Операция умножения определена для двух матриц, таких что число столбцов первой равно числу строк второй.

Пусть матрицы $A$ и $B$ таковы, что $A \in \mathbb{R}^{n \times k}$ и $B \in \mathbb{R}^{k \times m}$.    
__Произведением__ матриц $A$ и $B$ называется матрица $C$, такая что
$$c_{ij} = \sum_{r=1}^{k} a_{ir}b_{rj}$$,
где  $c_{ij}$ — элемент матрицы $C$, стоящий на пересечении строки с номером $i$ и столбца с номером $j$.

````python
a = np.array([[1, 2], [0, 1]])
b = np.array([[4, 1], [2, 2]])
print(a @ b)    # python 3 style
print(a.dot(b))
print(np.dot(a, b))
````
