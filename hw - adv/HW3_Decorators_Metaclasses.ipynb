```python
# ячейка с тестами


import gc

class Obj:
    def __init__(self, x):
        self.x = x

calls = 0

@lru_cache(3)
def f(o):
    global calls
    calls += 1
    return o.x * 10

# --- тесты ---

o1 = Obj(1)

assert f(o1) == 10
assert calls == 1
assert f(o1) == 10
assert calls == 1


# --- удаление weakref'а ---
del o1
gc.collect()

o1_new = Obj(1)

# нужно будет заново посчитать, т.к. мы уже удалили объект
assert f(o1_new) == 10
assert calls == 2

o2 = Obj(2)
o3 = Obj(3)

assert f(o2) == 20
assert f(o3) == 30
assert calls == 4
assert f(o2) == 20
assert calls == 4

o4 = Obj(4)

assert f(o4) == 40
assert calls == 5
o5 = Obj(5)
assert f(o5) == 50
assert calls == 6

assert f(o2) == 20
assert calls == 7

print("All tests passed")
```
